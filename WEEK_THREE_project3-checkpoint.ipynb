```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("Mall_Customers.csv")
print("First 5 Rows:")
df.head()
```

```python
print("Dataset Shape:", df.shape)
print("Dataset Info:")
df.info()
print("Statistical Summary:")
df.describe()
```

```python
print("Missing Values:")
df.isnull().sum()
```

```python
X = df[['Age', 'Spending Score (1-100)']]
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x='Age',
    y='Spending Score (1-100)',
    data=df,
    s=80
)
plt.title('Age vs Spending Score')
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.show()
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
wcss = []

for k in range(1,11):

    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    kmeans.fit(X_scaled)

    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1,11), wcss, marker='o')

plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')

plt.show()
```

```python
silhouette_scores = []

for k in range(2,11):

    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    labels = kmeans.fit_predict(X_scaled)

    score = silhouette_score(X_scaled, labels)

    silhouette_scores.append(score)

plt.figure(figsize=(8,5))
plt.plot(range(2,11), silhouette_scores, marker='o')

plt.title('Silhouette Scores')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')

plt.show()

best_k = np.argmax(silhouette_scores) + 2

print("Best Number of Clusters =", best_k)
```

```python
kmeans = KMeans(
    n_clusters=best_k,
    random_state=42,
    n_init=10
)

df['Cluster'] = kmeans.fit_predict(X_scaled)

print("\nCluster Assigned Successfully!")
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x='Age',
    y='Spending Score (1-100)',
    hue='Cluster',
    palette='Set1',
    data=df,
    s=100
)

plt.scatter(
    kmeans.cluster_centers_[:,0] * X['Age'].std() + X['Age'].mean(),
    kmeans.cluster_centers_[:,1] * X['Spending Score (1-100)'].std() + X['Spending Score (1-100)'].mean(),
    s=300,
    c='black',
    marker='X',
    label='Centroids'
)

plt.title('Customer Segmentation using K-Means')
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.legend()

plt.show()
```

```python
print("\nCluster Summary:")

cluster_summary = df.groupby('Cluster')[
    ['Age', 'Spending Score (1-100)']
].mean()

display(cluster_summary)
```

```python
print("Customers in Each Cluster:")
display(df['Cluster'].value_counts())
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
print("Explained Variance Ratio:")
print(pca.explained_variance_ratio_)
```

```python
pca_df = pd.DataFrame(
    X_pca,
    columns=['PC1', 'PC2'])
pca_df['Cluster'] = df['Cluster']
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x='PC1',
    y='PC2',
    hue='Cluster',
    palette='Set1',
    data=pca_df,
    s=100
)

plt.title('Customer Segmentation using PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.show()
```

```python
centroids_pca = pca.transform(
    kmeans.cluster_centers_
)

plt.figure(figsize=(10,6))

sns.scatterplot(
    x='PC1',
    y='PC2',
    hue='Cluster',
    palette='Set1',
    data=pca_df,
    s=100
)

plt.scatter(
    centroids_pca[:,0],
    centroids_pca[:,1],
    s=300,
    c='black',
    marker='X',
    label='Centroids'
)

plt.title('K-Means Clusters with PCA')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()

plt.show()
```

```python
variance_df = pd.DataFrame({
    'Principal Component': ['PC1', 'PC2'],
    'Explained Variance Ratio': pca.explained_variance_ratio_
})

print("\nPCA Explained Variance:")
display(variance_df)
```
